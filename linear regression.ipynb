```python
import pandas as pd
import numpy as np 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt 

#to check if dataset is linear
data=pd.read_csv("/kaggle/input/house-price-predection/house price.csv")
data.plot(kind='scatter',x="sqft_living",y="price")
plt.show()

#training
x_train,x_test,y_train,y_test = train_test_split(data["sqft_living"],data["price"])
plt.scatter(x_train,y_train,color="r")
plt.scatter(x_test,y_test,color="b")
plt.show()

#model
lr=LinearRegression()
lr.fit(x_train.values.reshape(-1,1),y_train.values)

#prediction
prediction=lr.predict(x_test.values.reshape(-1,1))
plt.plot(x_test,prediction)
plt.scatter(x_test,y_test)
plt.show()
sqft_to_predict =1000
print(lr.predict(np.array([[sqft_to_predict]]))[0])

#scoring
print(lr.score(x_test.values.reshape(-1,1),y_test.values))
```
